```python
df['Age'].describe()
```

```python
def age_bin(x):
  if x<=20:
    return "18-20"
  elif x>20 and x<=25:
    return "21-25"
  elif x>25 and x<=30:
    return "26-30"
  elif x>30 and x<=35:
    return "31-35"
  else:
    return "35+"
```

```python
df['Age_bin'] = df['Age'].apply(lambda x: age_bin(x))
```

```python
pd.crosstab(index = df['Age_bin'], columns = df['Product'])
```

- P(Product | Age_Bin)
- Divide the above crosstab with index

```python
pd.crosstab(index = df['Age_bin'], columns = df['Product'], normalize ='index')
```

- p(KP281 | 31-35) = 0.34
- P(KP481 | 31-35) = 0.53
- P(KP781 | 31-35) = 0.12

If we know that the customer age is between 31-35, we would offer the customer with KP481 as the customer is highly likely to buy it.
